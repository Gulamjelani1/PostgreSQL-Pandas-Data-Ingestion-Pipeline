```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.engine import URL
from dotenv import load_dotenv
import os
```

```python
# Load environment variables from .env
load_dotenv()

DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")

```

```python
# --- Legacy approach: psycopg2 ---
url_legacy = URL.create(
    drivername="postgresql+psycopg2",
    username=DB_USER,
    password=DB_PASSWORD,
    host=DB_HOST,
    port=DB_PORT,
    database=DB_NAME
)
```

```python
engine_legacy = create_engine(url_legacy)
```

```python
df_legacy = pd.read_sql("SELECT * FROM employees", engine_legacy)
```

```python
print(df_legacy)
```

```python
# --- Modern approach: psycopg3 ---
url_modern = URL.create(
    drivername="postgresql+psycopg",
    username=DB_USER,
    password=DB_PASSWORD,
    host=DB_HOST,
    port=DB_PORT,
    database=DB_NAME
)
```

```python
engine_modern = create_engine(url_modern)
```

```python
df_modern = pd.read_sql("SELECT * FROM employees", engine_modern)
```

```python
print(df_modern)
```
